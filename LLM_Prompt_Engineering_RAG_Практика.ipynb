## вычисление косинусных и евклидовых расстояний

```python
from sklearn.metrics.pairwise import cosine_distances, euclidean_distances

v1 = [[1, 2]]
v2 = [[5, 5]]

print("euclidean_distances", *euclidean_distances(v1, v2)[0])
print("cosine_distances", *cosine_distances(v1, v2)[0])

# euclidean_distances 5.0
# cosine_distances 0.05131670194948634
```

##3. Развертывание локальной модели эмбеддингов

**Задание:**
- Использовать модель `sentence-transformers/all-MiniLM-L6-v2` с помощью библиотеки Python sentence_transformers.
- Получить эмбеддинги для текста:
  - `"Привет, Мир!"`
  - `"Привет, ПММ!"`
- Рассчитать косинусное сходство между эмбеддингами.

**Ожидаемый результат:**
- Скрипт с результатами эмбеддингов и небольшое сравнение по косинусной мере.

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Получение эмбеддингов для текста "Привет, Мир!", "Привет, ПММ!"
sentences = ["Привет, Мир!", "Привет, ПММ!"]
embeddings = model.encode(sentences)

from sklearn.metrics.pairwise import cosine_distances
#Рассчет косинусного сходства между эмбеддингами.
cosine_similarity_matrix = 1. - cosine_distances(embeddings)

print(f"cosine_similarity_matrix:\n{cosine_similarity_matrix}")
```

```python
print(f"embeddings:\n{embeddings}")
```

Вывод:
*   Получены эмбеддинги для вышеуказанного текста
*   Была расчитана матрица сходства по косинусной мере, тексты "Привет, Мир!"
"Привет, ПММ!" являются схожими по косинусной мере, так как ее значение 0.85 приближенно к 1
