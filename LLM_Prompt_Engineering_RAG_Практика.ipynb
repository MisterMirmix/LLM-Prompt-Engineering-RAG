## 3. Развертывание локальной модели эмбеддингов

**Задание:**
- Использовать модель `sentence-transformers/all-MiniLM-L6-v2` с помощью библиотеки Python sentence_transformers.
- Получить эмбеддинги для текста:
  - `"Привет, Мир!"`
  - `"Привет, ПММ!"`
- Рассчитать косинусное сходство между эмбеддингами.

**Ожидаемый результат:**
- Скрипт с результатами эмбеддингов и небольшое сравнение по косинусной мере.

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Получение эмбеддингов для текста "Привет, Мир!", "Привет, ПММ!"
sentences = ["Привет, Мир!", "Привет, ПММ!"]
embeddings = model.encode(sentences)

from sklearn.metrics.pairwise import cosine_distances
#Рассчет косинусного сходства между эмбеддингами.
cosine_similarity_matrix = 1. - cosine_distances(embeddings)

print(f"cosine_similarity_matrix:\n{cosine_similarity_matrix}")
```

```python
print(f"embeddings:\n{embeddings}")
```

Вывод:
*   Получены эмбеддинги для вышеуказанного текста
*   Была расчитана матрица сходства по косинусной мере, тексты "Привет, Мир!"
"Привет, ПММ!" являются схожими по косинусной мере, так как ее значение 0.85 приближенно к 1

###Разделение текста на чанки с использованием разных моделей langchain_text_splitters

####RecursiveCharacterTextSplitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter_recursive = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap  = 20,
    length_function = len,
)
texts_recursive = text_splitter_recursive.create_documents([text_of_task_4])
print(texts_recursive[0])
print(texts_recursive[1])
```

####CharacterTextSplitter

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter_character = CharacterTextSplitter(
    separator = "\n\n",
    chunk_size = 1000,
    chunk_overlap  = 200,
    length_function = len,
)
texts_character = text_splitter_character.create_documents([text_of_task_4])
print(texts_character[0])
print(texts_character[1])
```

####GPT2TokenizerFast (HuggingFace Length Function)




```python
from transformers import GPT2TokenizerFast

tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
text_splitter_huggingface = CharacterTextSplitter.from_huggingface_tokenizer(tokenizer, chunk_size=100, chunk_overlap=0)
texts_huggingface = text_splitter_huggingface.split_text(text_of_task_4)
```

```python
print(texts_huggingface[0])
print(texts_huggingface[1])
print(texts_huggingface[2])
```

####tiktoken_encoder (tiktoken (OpenAI) Length Function)




```python
text_splitter_tiktoken = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=100, chunk_overlap=0)
texts_tiktoken = text_splitter_tiktoken.split_text(text_of_task_4)
```

```python
print(texts_tiktoken[0])
print(texts_tiktoken[1])
```
